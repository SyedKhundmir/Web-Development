```python
class Node:
    def __init__(self,data = None,next = None):
        self.data = data
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def append_node(self,new_data):
        new_node = Node(new_data)
        
        
        if self.head == None:
            self.head = new_node
        else:
            temp = self.head
            while(temp.next!=None):
                temp = temp.next
            temp.next = new_node
            
    
    def add_at_start(self,new_data):
        new_node = Node(new_data)
        
        if self.head == None:
            self.head = new-node
        
        else:
            temp = new_node
            temp.next = self.head
            self.head = temp
    
    def add_at_index(self,new_data,index):
        new_node = Node(new_data)
        if self.head == None:
            self.head = new_node
        else:
            if index == 0:
                temp = new_node
                temp.next = self.head
                self.head = temp
            else:
                count = 1
                temp = new_node
                forward = self.head
                while(count < index and forward.next != None):
                    forward = forward.next
                    count = count + 1
                temp2 = forward.next
                forward.next = temp
                temp.next = temp2
        
    def delete_node(self,del_data):
        if self.head == None:
            print("Empty list")
        flag = False
        temp = self.head
        if temp.data == del_data:
            self.head = temp.next
            temp.next = None
            flag = True
        else:
            temp = self.head
            forward = temp.next
            backward = temp
            
            while(forward.next != None):
                if forward.data == del_data:
                    backward.next = forward.next
                    forward.data = None
                    flag = True
                    break
                forward = forward.next
                backward = backward.next
                
            if forward.data == del_data:
                backward.next = None
                flag = True
                
        if flag == False:
            print("Element",del_data,"not present in list")
        else:
            print("Element",del_data,"is deleted")
            
    
    
    def display(self):
        temp = self.head
        
        while(temp.next != None):
            print(temp.data,end = "->")
            temp = temp.next
        print(temp.data)
        
    
    def count_nodes(self):
        if self.head == None:
            print("Empty list")
            
        else:
            count = 1
            temp = self.head
            while(temp.next != None):
                count = count + 1
                temp = temp.next
            print(count)
            
    def reverse_list(self):
        temp = None
        temp2 = None
        while(self.head != None):
            temp2 = self.head.next
            self.head.next = temp
            temp = self.head
            self.head = temp2
        self.head = temp
    
    
    def rotate_list(self):
        temp2 = self.head.next
        temp = self.head
        while(temp2.next != None):
            temp = temp.next
            temp2 = temp2.next
        temp2.next = self.head
        temp.next = None
        self.head = temp2
    
    def detect_loop(self):
        dic = {}
        flag = False
        temp = self.head
        while(temp.next != None):
            if temp not in dic:
                dic[temp] = temp.data
                temp = temp.next
            else:
                flag = True
                break
        if flag == True:
            print("Loop detected")
        else:
            print("No Loop")
```

```python
llist = LinkedList()
llist.append_node(10)
llist.append_node(20)
llist.append_node(30)
llist.append_node(40)
llist.append_node(50)
llist.display()

# Create a loop for testing
llist.head.next.next.next.next = llist.head
llist.detect_loop()
```
